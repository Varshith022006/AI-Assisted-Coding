```python
def is_even(n):
    if not isinstance(n, int):
        raise TypeError("Input must be an integer")
    return n % 2 == 0
```

```python
try:
    user_input = int(input("Enter an integer number: "))

    if is_even(user_input):
        print("True")
    else:
        print("False")

except ValueError:
    print("Error: Please enter a valid integer.")

except TypeError as e:
    print(e)
```

```python
def to_uppercase(text):
    if not isinstance(text, str):
        raise TypeError("Input must be a string")
    return text.upper()


def to_lowercase(text):
    if not isinstance(text, str):
        raise TypeError("Input must be a string")
    return text.lower()
```

```python
try:
    user_text = input("Enter a string: ")

    choice = input("Choose conversion (upper/lower): ").strip().lower()

    if choice == "upper":
        print(to_uppercase(user_text))
    elif choice == "lower":
        print(to_lowercase(user_text))
    else:
        print("Invalid choice. Please select 'upper' or 'lower'.")

except TypeError as e:
    print("Error:", e)
```

```python
def sum_list(numbers):
    total = 0
    for item in numbers:
        if isinstance(item, (int, float)):
            total += item
    return total
```

```python
user_input = input("Enter list elements separated by commas: ")

# If user enters nothing â†’ empty list
if user_input.strip() == "":
    data = []
else:
    raw_items = user_input.split(",")
    data = []

    for item in raw_items:
        item = item.strip()
        try:
            data.append(int(item))
        except ValueError:
            try:
                data.append(float(item))
            except ValueError:
                data.append(item)  # non-numeric safely ignored

result = sum_list(data)
print("Sum of list:", result)
```

```python
class StudentResult:

    def __init__(self):
        self.marks = []

    def add_marks(self, mark):
        if not isinstance(mark, (int, float)):
            raise ValueError("Marks must be numeric")
        if mark < 0 or mark > 100:
            raise ValueError("Marks must be between 0 and 100")
        self.marks.append(mark)

    def calculate_average(self):
        if len(self.marks) == 0:
            return 0
        return sum(self.marks) / len(self.marks)

    def get_result(self):
        if self.calculate_average() >= 40:
            return "Pass"
        return "Fail"
```

```python
student = StudentResult()

print("Enter student marks (type 'done' to finish):")

while True:
    value = input("Enter mark: ").strip()

    if value.lower() == "done":
        break

    try:
        mark = float(value)
        student.add_marks(mark)
    except ValueError as e:
        print("Error:", e)

average = student.calculate_average()
result = student.get_result()

print("\nAverage:", average)
print("Result:", result)
```

```python
def is_valid_username(username):
    if not isinstance(username, str):
        return False

    if len(username) < 5:
        return False

    if " " in username:
        return False

    if not username.isalnum():
        return False

    return True
```

```python
username = input("Enter a username: ")

if is_valid_username(username):
    print("Valid Username")
else:
    print("Invalid Username")

```
