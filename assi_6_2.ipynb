```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def is_pass(self):
        return self.marks >= 40


# Dynamic Input
name = input("Enter student name: ")
roll_no = int(input("Enter roll number: "))
marks = float(input("Enter marks: "))

# Object Creation
student = Student(name, roll_no, marks)

# Output
if student.is_pass():
    print("Result: PASS")
else:
    print("Result: FAIL")
```

```python
def pattern_while(n):
    i = 1
    while i <= n:
        print("*" * i)
        i += 1

pattern_while(5)
```

```python
def check_number(num):
    if num > 0:
        return "Positive"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"


# Dynamic Input
num = float(input("Enter a number: "))

# Function Call and Output
result = check_number(num)
print("The number is:", result)
```

```python
def check_discount(age, is_member):
    if age >= 60:
        print("Eligible for senior discount")
        if is_member:
            print("Eligible for additional member discount")
    else:
        if is_member:
            print("Eligible for member discount")
        else:
            print("No discount available")


# Dynamic Input
age = int(input("Enter age: "))
member_input = input("Are you a member? (yes/no): ").lower()

# Convert to boolean
is_member = True if member_input == "yes" else False

# Function Call
check_discount(age, is_member)
```

```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def circumference(self):
        return 2 * math.pi * self.radius


# Testing
c = Circle(7)
print("Area:", c.area())
print("Circumference:", c.circumference())
```
